Let's define the inputs of the voltage and current instantaneous values

```python
# Import math Library
import math

# Inputs for instantaneous voltage
vmax = 150#float(input('Vmax amplitude (V): '))
vangle = 10#float(input('v phase-angle (°): '))
vrms =  vmax/math.sqrt(2)
print('v(t)=',vmax,'*cos(w*t+',vangle,')')
print('Vrms (Volts):', round(vrms,2))

# Inputs for instantaneous current
imax = 5#float(input('Imax amplitude (A): '))
iangle = -50# float(input('i phase-angle (°): '))
irms =  imax/math.sqrt(2)
print('i(t)=',imax,'*cos(w*t+',iangle,')')
print('Irms (Volts):', round(irms,2))
```

a). To draw the power triangle it is required to determine S with the provided info, and subsequently determine P and Q.

```python
# Determine S
Smag=vrms*irms
Sang=abs(vangle)+abs(iangle)
print('Real power magnitude (VA):', round(Smag,2))
print('Real power angle (degrees):',Sang)

# Determine P
Pmag = Smag*math.cos(math.radians(Sang))
print('Active power magnitude (W):', round(Pmag,2))

# Determine Q
Qmag = Smag*math.sin(math.radians(Sang))
print('Reactive power magnitude (VAR):', round(Qmag,2))
```

```python
import matplotlib.pyplot as plt

# Vertices of the triangle
x = [0, Pmag, Pmag]
y = [0, 0, Qmag]

# Draw the triangle
plt.fill(x, y, 'r')

# Add text to the plot
plt.text(Pmag/4, Qmag/2, 'c: Real Power (VA)',fontdict={'family': 'serif', 'color': 'red', 'size': 10, 'weight': 'normal'})
plt.xlabel('b: Active Power (W)')
plt.ylabel('a: Reactive Power (VAR)')
plt.title('Power Triangle')
plt.title('Power Triangle')
```

b). To determine the power factor it is sufficient to use the mathemathical relationship pf = P/S

```python
# Determine the power factor
pf = Pmag/Smag
print('Power factor is:', round(pf,3), ' and it is Lagging')
```

c). To determine the reactive power from parallel capacitors to correct the pf to 0.9 it is needed to do the following.
1. Determine new reactive power of the load given the new provided power factor
2. Calculate the difference the new reactive power against the previous reactive power determined for the load

```python
# Determine the new Reactive power of the load
Qcorrected = Pmag*math.tan(math.acos(0.9))

# Determine the reactive power of the parallel capacitors
Qcap = Qmag-Qcorrected

print('New load reactive power (VAR):', round(Qcorrected,2))
print('Parallel Capacitors Supplied reactive power (VAR):', round(Qcap,2))
```
