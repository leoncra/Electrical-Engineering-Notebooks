a). Let's define the loads parameters and plot their respective power triangles.

```python
# Import Library
import math
import matplotlib.pyplot as plt

# Inputs for load 1
P1mag = 15000#float(input('Load 1 active power magnitude (W): '))
pf1 = 0.6#float(input('Load 1 power factor: '))
S1mag = P1mag/pf1
Q1mag = P1mag*math.tan(math.acos(pf1))
print('Load 1 Real power magnitude (VA):', round(S1mag,2))
print('Load 1 Real power angle (°):',round(math.degrees(math.acos(pf1)), 2))
print('Load 1 Active power magnitude (W):', round(P1mag,2))
print('Load 1 Reactive power magnitude (VAR):', round(Q1mag,2))
print('Load 1 Power factor is:', round(pf1,3), ' and it is Lagging')
print(10*'-')

# Inputs for load 2
S2mag = 10000#¤float(input('Load 2 Real power magnitude (VA): '))
pf2 = 0.8#float(input('Load 2 power factor: '))
P2mag = S2mag*pf2
Q2mag = P2mag*math.tan(math.acos(pf2))
print('Load 2 Real power magnitude (VA):', round(S2mag,2))
print('Load 2 Real power angle (°):',round(math.degrees(math.acos(pf2)), 2))
print('Load 2 Active power magnitude (W):', round(P2mag,2))
print('Load 2 Reactive power magnitude (VAR):', round(Q2mag,2))
print('Load 2 Power factor is:', round(pf2,3), ' and it is Leading')
```

```python
# load 1 plot 

# Vertices of the triangle 
x1 = [0, P1mag, P1mag]
y1 = [0, 0, Q1mag]

# Draw the triangle
plt.fill(x1, y1, 'r')

# Add text to the plot
plt.text(P1mag/5, Q1mag/2, 'c: Real Power (VA)',fontdict={'family': 'serif', 'color': 'red', 'size': 10, 'weight': 'normal'})
plt.xlabel('b: Active Power (W)')
plt.ylabel('a: Reactive Power (VAR)')
plt.title('Power Triangle Load 1')
```

```python
# load 2 plot 

# Vertices of the triangle 
x2 = [0, P2mag, P2mag]
y2 = [0, 0, -Q2mag]

# Draw the triangle
plt.fill(x2, y2, 'b')

# Add text to the plot
plt.text(P2mag/5, -Q2mag/2, 'c: Real Power (VA)',fontdict={'family': 'serif', 'color': 'blue', 'size': 10, 'weight': 'normal'})
plt.xlabel('b: Active Power (W)')
plt.ylabel('a: Reactive Power (VAR)')
plt.title('Power Triangle Load 1')
```

```python
# Combined load parameters
Pcmag = P1mag+P2mag
Qcmag = Q1mag-Q2mag
Scmag = math.sqrt(Pcmag**2 + Qcmag**2)
pfc = Pcmag/Scmag
print('Combined Load Real power magnitude (VA):', round(Scmag,2))
print('Combined Load Real power angle (°):',round(math.degrees(math.acos(pfc)), 2))
print('Combined Load Active power magnitude (W):', round(Pcmag,2))
print('Combined Load Reactive power magnitude (VAR):', round(Qcmag,2))
print('Combined Load Power factor is:', round(pfc,3), ' and it is Lagging')

# Combined load plot 

# Vertices of the triangle 
xc = [0, Pcmag, Pcmag]
yc = [0, 0, Qcmag]

# Draw the triangle
plt.fill(xc, yc, 'g')

# Add text to the plot
plt.text(Pcmag/5, Qcmag/2, 'c: Real Power (VA)',fontdict={'family': 'serif', 'color': 'green', 'size': 10, 'weight': 'normal'})
plt.xlabel('b: Active Power (W)')
plt.ylabel('a: Reactive Power (VAR)')
plt.title('Power Triangle Load 1')
```
